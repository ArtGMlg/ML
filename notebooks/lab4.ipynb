```python
import numpy as np

def cross_entropy(y_pred, y_true):
    """
    Рассчитывает кросс-энтропию между двумя распределениями.

    Аргументы:
    y_true -- массив NumPy, содержащий фактические метки классов
    y_pred -- массив NumPy, содержащий вероятность принадлежности каждому классу

    Возвращает:
    Значение кросс-энтропии между y_true и y_pred
    """

    # Преобразуем метки классов в one-hot-кодировку
    y_true_onehot = np.zeros(y_pred.T.shape)
    y_true_onehot[np.arange(len(y_true)), y_true] = 1

    # Рассчитываем кросс-энтропию между y_true_onehot и y_pred
    return -np.sum(np.multiply(y_true_onehot, np.log(y_pred)))
```
